3. Em quais anos houve maiores quedas e valoriza√ß√µes?;


```python
# Extrair o ano da coluna de data para cada criptomoeda
bitcoin['Year'] = pd.to_datetime(bitcoin['Date']).dt.year
ethereum['Year'] = pd.to_datetime(ethereum['Date']).dt.year
cardano['Year'] = pd.to_datetime(cardano['Date']).dt.year


# Calcular varia√ß√£o percentual dos valores de fechamento em rela√ß√£o ao ano anterior para cada criptomoeda
bitcoin['Variation'] = bitcoin['Close'].pct_change()
ethereum['Variation'] = ethereum['Close'].pct_change()
cardano['Variation'] = cardano['Close'].pct_change()

# Agrupar por ano e encontrar a maior varia√ß√£o positiva (alta) e a menor varia√ß√£o negativa (queda) para cada criptomoeda
bitcoin_max_increase = bitcoin.groupby('Year')['Variation'].max()
bitcoin_min_decrease = bitcoin.groupby('Year')['Variation'].min()
ethereum_max_increase = ethereum.groupby('Year')['Variation'].max()
ethereum_min_decrease = ethereum.groupby('Year')['Variation'].min()
cardano_max_increase = cardano.groupby('Year')['Variation'].max()
cardano_min_decrease = cardano.groupby('Year')['Variation'].min()


# Identificar os anos com as maiores altas e quedas para cada criptomoeda
bitcoin_max = bitcoin_max_increase[bitcoin_max_increase == bitcoin_max_increase.max()].index
bitcoin_min = bitcoin_min_decrease[bitcoin_min_decrease == bitcoin_min_decrease.min()].index
ethereum_max = ethereum_max_increase[ethereum_max_increase == ethereum_max_increase.max()].index
ethereum_min = ethereum_min_decrease[ethereum_min_decrease == ethereum_min_decrease.min()].index
cardano_max = cardano_max_increase[cardano_max_increase == cardano_max_increase.max()].index
cardano_min = cardano_min_decrease[cardano_min_decrease == cardano_min_decrease.min()].index

# Converter os anos para listas
bitcoin_max = bitcoin_max.tolist()
bitcoin_min = bitcoin_min.tolist()
ethereum_max = ethereum_max.tolist()
ethereum_min = ethereum_min.tolist()
cardano_max = cardano_max.tolist()
cardano_min = cardano_min.tolist()


print(f'Bitcoin - Maiores Altas: {bitcoin_max}\nBitcoin - Maiores Quedas: {bitcoin_min}\nEthereum - Maiores Altas: {ethereum_max}\nEthereum - Maiores Quedas: {ethereum_min}\nCardano - Maiores Altas: {cardano_max}\nCardano - Maiores Quedas: {cardano_min}')
```

5/6. Qual moeda se mostra mais e menos interessante em rela√ß√£o √† valoriza√ß√£o pela an√°lise da s√©rie hist√≥rica?;


```python
#Pedi pro chatgpt explicar como eu faria essa avalia√ß√£o. A explica√ß√£o dele foi o texto abaixo:

'''Para avaliar qual moeda pode ser mais interessante para investir com base na valoriza√ß√£o hist√≥rica,
voc√™ pode calcular o retorno total de cada moeda ao longo do per√≠odo analisado.
O retorno total √© a diferen√ßa percentual entre o valor inicial e o valor final da moeda.

Nesse exemplo, utilizamos o valor de fechamento inicial e final de cada moeda para calcular o retorno total.
O retorno total √© a diferen√ßa percentual entre o valor final e o valor inicial, multiplicado por 100 para obter o valor percentual.

Ao realizar essas etapas, voc√™ obter√° o retorno total para cada moeda, representando a valoriza√ß√£o ao longo do per√≠odo analisado.
O valor percentual positivo indica uma valoriza√ß√£o, enquanto o valor percentual negativo indica uma desvaloriza√ß√£o.
'''

# Mapear IDs aos nomes das criptomoedas
nomes_criptomoedas = {
    0: 'Bitcoin',
    1: 'Ethereum',
    2: 'Cardano',
    3: 'Dash',
    4: 'Bitcoin Cash',
    5: 'IOTA',
    6: 'Litecoin',
    7: 'Monero',
    8: 'NEM',
    9: 'Ripple'
}

# Calcular o retorno total para cada moeda
returns = {}

for moeda in lista2:
    symbol = int(moeda['ID'].unique()[0])  # Converter para inteiro
    close = moeda['Close']
    return_total = (close.iloc[-1] - close.iloc[0]) / close.iloc[0] * 100
    nome_moeda = nomes_criptomoedas[symbol]
    returns[nome_moeda] = return_total

# Ordenar as moedas com base no retorno total em ordem decrescente
sorted_returns = sorted(returns.items(), key=lambda x: x[1], reverse=True)

# Imprimir os retornos totais de cada moeda em ordem decrescente
for symbol, return_total in sorted_returns:
    print(f'Retorno Total {symbol}: {return_total:.2f}%')
```

Observando a lista podemos perceber que a que a menos interessante se torna a NEM e a mais sendo o ETH

7. Existe correla√ß√£o entre os valores para todas as criptomoedas?

```python
# Calcular a correla√ß√£o entre as colunas 'Close' das criptomoedas
correlacao = pd.concat([bitcoin['Close'], ethereum['Close'], cardano['Close'], btccash['Close'], dash['Close'], iota['Close'], litecoin['Close'], monero['Close'], nem['Close'], ripple['Close']], axis=1).corr()

# Exibir a matriz de correla√ß√£o
correlacao
```

```python
# Calcular a matriz de correla√ß√£o entre as colunas 'Close' das criptomoedas
correlation_matrix = pd.concat([bitcoin['Close'], ethereum['Close'], cardano['Close'], btccash['Close'], dash['Close'], iota['Close'], litecoin['Close'], monero['Close'], nem['Close'], ripple['Close']], axis=1).corr()

# Criar o DataFrame personalizado para o mapa de calor com as etiquetas de eixo
labels = ['Bitcoin', 'Ethereum', 'Cardano', 'BTCCash', 'Dash', 'Iota', 'Litecoin', 'Monero', 'Nem', 'Ripple']
custom_corr_matrix = pd.DataFrame(correlation_matrix.values, index=labels, columns=labels)

# Criar o mapa de calor
ax = sns.heatmap(custom_corr_matrix, annot=True, cmap='coolwarm')

# Configurar o gr√°fico
plt.title('Correla√ß√£o entre Criptomoedas')
plt.xlabel('Criptomoedas')
plt.ylabel('Criptomoedas')

# Exibir o gr√°fico
plt.show()
```
